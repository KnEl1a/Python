```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FuncFormatter  # Importa FuncFormatter

#funciones para los formatos de los ejes
def millions(x, pos):
    'El número 1 se representa como 1M'
    return f'{x/1e6:.0f}M'

def millions_decimals(x, pos):
    'El número 1 se representa como 1M'
    return f'{x/1e6:.2f}M'

def porc(x, pos):
    'El número se representa como 1%'
    return f'{x/1:.0f}%'
    
plt.figure(figsize=(10, 5))
# Crea un gráfico de líneas con tres ejes Y
fig, ax1 = plt.subplots()

# Crear las barras verticales apiladas

c = ax1.bar(df['Unnamed: 0'], df['BRASIL'], label="Brasil", color="green", alpha=0.82)
b = ax1.bar(df['Unnamed: 0'], df['ARG'], label="Argentina", color="lightblue", alpha=0.82)
a = ax1.bar(df['Unnamed: 0'], df['URUGUAY'], label="Uruguay", color="yellow", alpha=0.82)
ax1.set_ylabel('Toneladas', color='Black')

# Eje Y 2
ax2 = ax1.twinx()
#ax2.set_ylabel('%', color='tab:blue')
ax2.set_ylabel('% Brasil', color='green')
ax2.plot(df['Unnamed: 0'], df['%'], color='#006400', label='Brasil', marker=">", markersize=3, linestyle = "--")


ax2.tick_params(axis='y', labelcolor='tab:green')

# Eje Y 3
ax3 = ax1.twinx() # compartimos grafico con ax1 pero usamos otro eje
ax3.spines['right'].set_position(('outward', 60)) # spines se refiere a los bordes del área de trazado (spines) en el gráfico.
ax3.set_ylabel('% Argentina y Uruguay', color='blue')
ax3.plot(df['Unnamed: 0'], df['%.2'], color='yellow', label='Uruguay', marker=">", markersize=3, linestyle = "-.")
ax3.plot(df['Unnamed: 0'], df['%.1'], color='tab:blue', label='Argentina', marker=">", markersize=3, linestyle = "-.")
ax3.tick_params(axis='y', labelcolor='blue')

# Añadir etiquetas en las barras


# Añadir leyenda y etiquetas de los ejes
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.07), ncol=4)
#ax1.legend(loc='upper center', bbox_to_anchor=(0.5, 0), ncol=4)

#plt.xlabel("Año")
#plt.ylabel("Tn")


# Configura el eje Y para mostrar los valores completos
ax1.yaxis.set_major_formatter(FuncFormatter(millions))
ax2.yaxis.set_major_formatter(FuncFormatter(porc))
ax3.yaxis.set_major_formatter(FuncFormatter(porc))

#plt.title("Religiosidad en Sudamérica")

ax1.bar_label(a, label_type='edge',fmt=FuncFormatter(millions_decimals),color = 'black')
ax1.bar_label(b, label_type='edge',fmt=FuncFormatter(millions_decimals),padding = 7, color = 'black')
ax1.bar_label(c, label_type='edge', fmt=FuncFormatter(millions_decimals),padding = 0.8,color = 'black')


#ax1.grid(True, which='major', axis='y', color='black', linestyle='--', linewidth=0.3, alpha=0.2)
ax3.grid(True, which='major', axis='y', color='green', linestyle='--', linewidth=0.5, alpha=0.3)
ax3.grid(True, which='major', axis='y', color='blue', linestyle='--', linewidth=0.5, alpha=0.3)
# Mostrar el gráfico
plt.savefig(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 30\graf.png', dpi=300, bbox_inches='tight')
plt.show()
```
