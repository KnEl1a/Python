```python
import numpy as np
import matplotlib.pyplot as plt

# Especies de pingüinos
species = ('URUGUAY', 'ARG', 'BRASIL')

# Conteo de pingüinos por sexo
sex_counts = df[['URUGUAY','ARG','BRASIL']]

# Ancho de las barras
width = 0.6

# Crear una figura y ejes
fig, ax = plt.subplots()

# Inicializar la posición inferior de las barras apiladas
bottom = np.zeros(3)

# Iterar sobre los conteos de pingüinos por sexo
for sex, sex_count in sex_counts.items():
    # Crear las barras para cada especie
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    
    # Actualizar la posición inferior para las próximas barras
    bottom += sex_count

    # Añadir etiquetas en el centro de cada barra
    ax.bar_label(p, label_type='center')

# Configurar el título del gráfico
ax.set_title('Number of penguins by sex')

# Añadir una leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def millions(x, pos):
    'El número 1 se representa como 1M'
    return f'{x/1e6:.0f}M'

plt.figure(figsize=(10, 5))

# Crear las barras verticales apiladas
bars = ['Uruguay', 'Argentina', 'Brasil']
bottom = np.zeros(len(df['Unnamed: 0']))  # Inicializar las alturas bases
for country in bars:
    plt.bar(df['Unnamed: 0'], df[country], label=country, bottom=bottom)
    bottom += df[country]

# Configurar el formato del eje Y para mostrar los valores completos
plt.gca().yaxis.set_major_formatter(millions)

# Añadir etiquetas en las barras
for bar in bars:
    plt.bar_label(plt.bar(df['Unnamed: 0'], df[bar], bottom=bottom), label_type='center', fmt='%.2f%%')

# Añadir leyenda y etiquetas de los ejes
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)
plt.xlabel("Año")
plt.ylabel("Porcentaje de participación en las exportaciones a China")

# Añadir una cuadrícula en el eje Y
plt.grid(True, which='major', axis='y', color='black', linestyle='--', linewidth=0.5, alpha=0.3)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FuncFormatter  # Importa FuncFormatter

def millions(x, pos):
    'El número 1 se representa como 1M'
    return f'{x/1e6:.0f}M'

def millions_decimals(x, pos):
    'El número 1 se representa como 1M'
    return f'{x/1e6:.2f}M'

plt.figure(figsize=(10, 5))
# Crea un gráfico de líneas con tres ejes Y
fig, ax1 = plt.subplots()

# Crear las barras verticales apiladas

c = ax1.bar(df['Unnamed: 0'], df['BRASIL'], label="Brasil", color="green")
b = ax1.bar(df['Unnamed: 0'], df['ARG'], label="Argentina", color="lightblue")
a = ax1.bar(df['Unnamed: 0'], df['URUGUAY'], label="Uruguay", color="red")

# Añadir etiquetas en las barras


# Añadir leyenda y etiquetas de los ejes
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=4)
#plt.xlabel("Año")
#plt.ylabel("Porcentaje de participación en las exportaciones a China")


# Configura el eje Y para mostrar los valores completos
ax1.yaxis.set_major_formatter(FuncFormatter(millions))

#plt.title("Religiosidad en Sudamérica")

plt.bar_label(a, label_type='edge',fmt=FuncFormatter(millions_decimals),color = 'red')
plt.bar_label(b, label_type='edge',fmt=FuncFormatter(millions_decimals), color = 'lightblue')
plt.bar_label(c, label_type='edge', fmt=FuncFormatter(millions_decimals),color = 'green')


plt.grid(True, which='major', axis='y', color='black', linestyle='--', linewidth=0.5, alpha=0.3)
# Mostrar el gráfico
plt.show()
```
