```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FuncFormatter  # Importa FuncFormatter
```

```python
df = pd.read_excel(r"C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf.31\Datos bs.xlsx")
df.head() # muestra la "cola" de la tabla, es decir,  final
```

```python
#asi da error #df[["cereales %"], ["oleaginosos %"]] = df[["cereales %"], ["oleaginosos %"]] * 100
df[["cereales %", "oleaginosos %"]] = df[["cereales %", "oleaginosos %"]] * 100
```

```python
#df[["cereales %", "oleaginosos %"]]=df[["cereales %", "oleaginosos %"]].astype(int)
df[["cereales %", "oleaginosos %"]] = df[["cereales %", "oleaginosos %"]].fillna(0)
df[["cereales %", "oleaginosos %"]] = df[["cereales %", "oleaginosos %"]].astype(int)
```

```python
# esta bien pero ahora para una mejor visualizacion vamos a hacer unos histogramas vreticales para los % 


plt.figure(figsize=(16, 8))
def millions(x, pos):
    'El número 1 se representa como 1M'
    return f'{x:.0f}M'

def k(x, pos):
    'El número 1 se representa como 1M'
    return f'{x:.0f}%'

# Crea un gráfico de líneas con tres ejes Y
fig, ax1 = plt.subplots()

# Eje Y 1
ax1.set_xlabel('Años')
ax1.set_ylabel('Exportaciones en millones de dolares', color='brown')
ax1.plot(df['Unnamed: 0'], df['cereales'], color='brown', label='Cereales', marker="<", markersize=3)
ax1.plot(df['Unnamed: 0'], df['oleaginosos'], color='tab:green', label='Oleaginosos', marker=">", markersize=3)
ax1.tick_params(axis='y', labelcolor='blue')


# Eje Y 2
ax2 = ax1.twinx()
ax2.set_ylabel('% de dichas exportaciones', color='tab:red')
#ax2.plot(df['Unnamed: 0'], df['Tn_B'], color='tab:red', label='Producción', marker=">", markersize=3, linestyle = "-.")
#ax2.plot(df['Unnamed: 0'], df['% de Participación china _a'], color='tab:red', label='% de china sobre Argentina', marker="*", markersize=3, linestyle = "--")
#ax2.plot(df['Unnamed: 0'], df['% de Participación china_b'], color='tab:orange', label='% de china sobre Brasil', marker="+", markersize=3, linestyle = "-.")

#anteriormente estos eran granficos de lineas y para una mejor claridad en el grafico los hacemos de barras verticales
#Ademas hacemos estos elementos algo mas transparentes con el parametro "alpha" 

plt.plt(df['Unnamed: 0'], df['cereales %'], label = "% de Cereales", color = "red", alpha=0.52)
plt.plt(df['Unnamed: 0'] + 0.2, df['oleaginosos %'], label = "% de Oleaginosos", color = "orange", alpha=0.52)


ax2.tick_params(axis='y', labelcolor='tab:red')


#activar leyenda con cuadro más pequeño
ax1.legend(loc='upper left', bbox_to_anchor=(0.001, 0.98), handlelength=1, fontsize='small')
ax2.legend(loc='upper left', bbox_to_anchor=(0.001, 0.86), handlelength=1, fontsize='small')

#ax3.legend(loc='upper left')

# Configura el eje Y para mostrar los valores completos
ax1.yaxis.set_major_formatter(FuncFormatter(millions))
ax2.yaxis.set_major_formatter(FuncFormatter(k))

# Especificar el formato de los ticks del eje X para mostrar los años cada 10 años
plt.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))

# Establecer los ticks del eje X manualmente cada 10 años
plt.xticks(np.arange(2000, 2019, 4),fontsize=3)

#activar cuadricula 
#plt.grid()

# Activar la cuadrícula en el primer eje (ax1)
#ax1.grid(True, axis='y', linestyle='-', color='gray', alpha=0.5)
plt.grid(True, which='major', axis='both', color='red', linestyle='--', linewidth=0.8, alpha=0.34)
ax1.grid(True, which='major', axis='both', color='blue', linestyle='--', linewidth=0.5, alpha=0.35)
#ax2.grid(True, which='major', axis='both', color='red', linestyle='--', linewidth=0.5, alpha=0.3)
#activar marcas menores 

plt.minorticks_on()

#plt.title("Segun la FAO(1961-2019): Área cosechada, rendimiento y producción en China del Maiz")
#plt.savefig(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 26\graf. lineas 2Axis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter, ScalarFormatter

# Funciones para formatear los ticks
def millions(x, pos):
    'El número 1 se representa como 1M'
    return f'{x:.0f}M'

def k(x, pos):
    'El número 1 se representa como 1M'
    return f'{x:.0f}%'

# Crear la figura y los ejes
fig, ax1 = plt.subplots(figsize=(16, 8))

# Eje Y 1
ax1.set_xlabel('Años')
ax1.set_ylabel('Exportaciones en millones de dólares', color='brown')
ax1.plot(df['Unnamed: 0'], df['cereales'], color='brown', label='Cereales', marker="<", markersize=3)
ax1.plot(df['Unnamed: 0'], df['oleaginosos'], color='tab:green', label='Oleaginosos', marker=">", markersize=3)
ax1.tick_params(axis='y', labelcolor='blue')

# Eje Y 2
ax2 = ax1.twinx()
ax2.set_ylabel('% del total de exportaciones Bonaerences', color='tab:red')

# Crear gráficos de barras verticales
ax2.bar(df['Unnamed: 0'] - 0.2, df['cereales %'], width=0.4, label='% de Cereales', color='red', alpha=0.52)
ax2.bar(df['Unnamed: 0'] + 0.2, df['oleaginosos %'], width=0.4, label='% de Oleaginosos', color='orange', alpha=0.52)

ax2.tick_params(axis='y', labelcolor='tab:red')

# Activar leyenda con cuadro más pequeño
ax1.legend(loc='upper left', bbox_to_anchor=(0.001, 0.98), handlelength=1, fontsize='small')
ax2.legend(loc='upper left', bbox_to_anchor=(0.001, 0.90), handlelength=1, fontsize='small')

# Configura el eje Y para mostrar los valores completos
ax1.yaxis.set_major_formatter(FuncFormatter(millions))
ax2.yaxis.set_major_formatter(FuncFormatter(k))

# Especificar el formato de los ticks del eje X para mostrar los años cada 10 años
plt.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))

# Establecer los ticks del eje X manualmente cada 4 años
plt.xticks(np.arange(2000, 2021, 1), fontsize=10)

# Activar la cuadrícula
plt.grid(True, which='major', axis='both', color='red', linestyle='--', linewidth=0.8, alpha=0.34)
ax1.grid(True, which='major', axis='both', color='blue', linestyle='--', linewidth=0.5, alpha=0.35)

# Activar marcas menores
plt.minorticks_on()

#plt.title("Exportaciones de Cereales y Oleaginosos y su Participación % (2000-2019)")
plt.savefig(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 31\graf. lineas BS.AS 2Axis.png', dpi=300, bbox_inches='tight')
plt.show()
```
