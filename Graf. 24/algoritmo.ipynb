```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FuncFormatter  # Importa FuncFormatter
```

```python
# aca modifique la ruta de su archivo xlsx para poder procesegui con el algoritmo, en mi caso mi archivo se encontraba en "descargas"
df_soja = pd.read_excel(r"C:\Users\Usuario\Downloads\EXP_Harina_de_soja_graf24.xlsx")
```

```python
df_soja[["ARGENTINA", "BRASIL", "IMPORTACIONES CHINAS TOTALES"]] = df_soja[["ARGENTINA", "BRASIL", "IMPORTACIONES CHINAS TOTALES"]].astype(int)
# convertimos el objeto actual de dichas columnas en tipo entero
```

```python
df_soja[["ARGENTINA", "BRASIL", "IMPORTACIONES CHINAS TOTALES"]]
```

```python
df_soja["Unnamed: 0"] = df_soja["Unnamed: 0"].astype(int)
df_soja.info()
```

```python
plt.figure(figsize=(16, 8))

# Crea un gráfico de líneas con tres ejes Y
fig, ax1 = plt.subplots()

# Eje Y 1
ax1.set_xlabel('Años')
ax1.set_ylabel('Toneladas', color='black')
ax1.plot(df_soja['Unnamed: 0'], df_soja['ARGENTINA'], marker='.', markersize=3, label = "Argentina", color = "blue")
ax1.plot(df_soja['Unnamed: 0'], df_soja['BRASIL'], marker='+', markersize=3, label = "Brasil", color = "green")
ax1.plot(df_soja['Unnamed: 0'], df_soja['IMPORTACIONES CHINAS TOTALES'], marker='*', markersize=3, label = "Importaciones Chinas", color = "red")

ax1.tick_params(axis='y', labelcolor='black')

#activar leyenda con cuadro más pequeño
ax1.legend(loc='upper left', bbox_to_anchor=(0.001, 0.97), handlelength=1)

# Especificar el formato de los ticks del eje X para mostrar los años cada 10 años
plt.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))

# Establecer los ticks del eje X manualmente cada 10 años
plt.xticks(np.arange(2000, 2019, 4),fontsize=9)
#plt.yticks(np.arange(210000, 2020, 3))

# Establecer los ticks del eje X manualmente cada 10 años
plt.yticks(np.arange(0, 700001, 50000),fontsize=9)

#activar cuadricula 
#plt.grid()

# Activar la cuadrícula en el primer eje (ax1)
#ax1.grid(True, axis='y', linestyle='-', color='gray', alpha=0.5)
plt.grid(True, which='major', axis='both', color='blue', linestyle='--', linewidth=0.8, alpha=0.5)
ax1.grid(True, which='major', axis='both', color='black', linestyle='--', linewidth=0.5, alpha=0.35)
#ax2.grid(True, which='major', axis='both', color='red', linestyle='--', linewidth=0.5, alpha=0.3)
#activar marcas menores 

plt.minorticks_on()

#plt.title("Gráfico de exportaciones de harina de soja de ARG y Brasil a China e importaciones totales de China (al mundo) en toneladas")
#plt.savefig(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 24\graf. 24 de lineas.png', dpi=300, bbox_inches='tight')
plt.show()
```
