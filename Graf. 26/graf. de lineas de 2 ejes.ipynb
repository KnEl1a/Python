```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FuncFormatter  # Importa FuncFormatter
```

```python
df = pd.read_excel(r"C:\Users\Usuario\Downloads\Exportaciones de ACEITE de soja - ARG y BRA al mundo y a CHINA.xlsx")
df.tail() # muestra la "cola" de la tabla, es decir,  final
```

```python
df.info()
```

```python
plt.figure(figsize=(16, 8))
def millions(x, pos):
    'El número 1 se representa como 1M'
    return f'{x/1e6:.0f}M'

def k(x, pos):
    'El número 1 se representa como 1M'
    return f'{x:.0f}%'

# Crea un gráfico de líneas con tres ejes Y
fig, ax1 = plt.subplots()

# Eje Y 1
ax1.set_xlabel('Años')
ax1.set_ylabel('Exportaciones en toneladas', color='blue')
ax1.plot(df['Unnamed: 0'], df['Tn'], color='blue', label='Argentina a China', marker="<", markersize=3)
ax1.plot(df['Unnamed: 0'], df['Tn_B'], color='tab:green', label='Brasil a China', marker=">", markersize=3)
ax1.tick_params(axis='y', labelcolor='blue')


# Eje Y 2
ax2 = ax1.twinx()
ax2.set_ylabel('% de Participación china sobre las exportaciones', color='tab:red')
#ax2.plot(df['Unnamed: 0'], df['Tn_B'], color='tab:red', label='Producción', marker=">", markersize=3, linestyle = "-.")
ax2.plot(df['Unnamed: 0'], df['% de Participación china _a'], color='tab:red', label='% de china sobre Argentina', marker="*", markersize=3, linestyle = "--")
ax2.plot(df['Unnamed: 0'], df['% de Participación china_b'], color='tab:orange', label='% de china sobre Brasil', marker="+", markersize=3, linestyle = "-.")
ax2.tick_params(axis='y', labelcolor='tab:red')


#activar leyenda con cuadro más pequeño
ax1.legend(loc='upper left', bbox_to_anchor=(0.001, 0.97), handlelength=1, fontsize='small')
ax2.legend(loc='upper left', bbox_to_anchor=(0.001, 0.90), handlelength=1, fontsize='small')

#ax3.legend(loc='upper left')

# Configura el eje Y para mostrar los valores completos
ax1.yaxis.set_major_formatter(FuncFormatter(millions))
ax2.yaxis.set_major_formatter(FuncFormatter(k))

# Especificar el formato de los ticks del eje X para mostrar los años cada 10 años
plt.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))

# Establecer los ticks del eje X manualmente cada 10 años
plt.xticks(np.arange(2000, 2019, 4),fontsize=3)

#activar cuadricula 
#plt.grid()

# Activar la cuadrícula en el primer eje (ax1)
#ax1.grid(True, axis='y', linestyle='-', color='gray', alpha=0.5)
plt.grid(True, which='major', axis='both', color='red', linestyle='--', linewidth=0.8, alpha=0.5)
ax1.grid(True, which='major', axis='both', color='blue', linestyle='--', linewidth=0.5, alpha=0.35)
#ax2.grid(True, which='major', axis='both', color='red', linestyle='--', linewidth=0.5, alpha=0.3)
#activar marcas menores 

plt.minorticks_on()

#plt.title("Segun la FAO(1961-2019): Área cosechada, rendimiento y producción en China del Maiz")
#plt.savefig(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 2\Correcion\graf. linea arroz.png', dpi=300, bbox_inches='tight')
plt.show()
```
