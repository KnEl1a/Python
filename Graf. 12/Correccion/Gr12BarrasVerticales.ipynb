```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv("principales-pases.csv", sep=';', decimal=',')

# Configurar la paleta de colores azul-celeste
custom_palette = sns.color_palette(["red"])

# Crear el gráfico de barras horizontales
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="% Participación", y="Category", data=data, palette="deep")

# Añadir título y etiquetas
plt.title('Principales destinos de las exportaciones del Mercosur ')
plt.xlabel('Participación FOB(mil millones de dolares)')
plt.ylabel('País')

# Agregar etiquetas con porcentaje dentro de cada barra
for p in ax.patches:
    ax.text(p.get_width() * 1.02, p.get_y() + p.get_height() / 2, '{:.0f}'.format(p.get_width()), ha='left', va='center')

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Cargar los datos desde el archivo CSV
data = pd.read_csv("principales-pases.csv", sep=';', decimal=',')

# Configurar la paleta de colores azul-celeste
custom_palette = sns.color_palette(["red"])

# Crear el gráfico de barras horizontales
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="% Participación", y="Category", data=data, palette="deep")

# Añadir título y etiquetas
plt.title('Principales destinos de las exportaciones del Mercosur en el 2019')
plt.xlabel('Participación FOB (mil millones de dólares)')
plt.ylabel('País')

# Cambiar el formato del eje x a mil millones de dólares
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}'.format(x/1e9)))

# Agregar etiquetas con porcentaje dentro de cada barra
for p in ax.patches:
    ax.text(p.get_width() * 1.02, p.get_y() + p.get_height() / 2, '{:.0f}'.format(p.get_width()), ha='left', va='center')

# Mostrar el gráfico y ...
# Guardar el gráfico en formato PNG con una resolución de 300 dpi
plt.savefig('grafico_exportaciones.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python

# Supongamos que ya tienes data definido , lo  dejamos asi

labels = data['Category']
sizes = data['% Participación']

# Ajustar el tamaño de la figura
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))

wedges, texts, autotexts = ax.pie(sizes, autopct='%1.1f%%', textprops=dict(color="w"), startangle=50,
                                  wedgeprops=dict(width=0.5, linewidth=0.4, edgecolor='black'), pctdistance=0.89)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"), bbox=bbox_props, zorder=0, va="center")

for i, (p, label) in enumerate(zip(wedges, labels)):
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f"angle,angleA=0,angleB={ang}"
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    
    # Ajustar la posición de las etiquetas y usar color de texto automático
    ax.annotate(label, xy=(x, y), xytext=(1.35 * np.sign(x), 1.4 * y),
                horizontalalignment=horizontalalignment, **kw, color=p.get_facecolor())

# Puedes personalizar la visualización según tus necesidades
# ax.set_title("Matplotlib bakery: A donut")
#plt.savefig(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 11\dona año2019.png', dpi=300, bbox_inches='tight')
plt.show()
```
