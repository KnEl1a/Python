```python
# r"C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 5\correcion\Wheat.csv"
import pandas as pd

# Lee el archivo Excel
datos = pd.read_excel(r"C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 4\Correcion\maiz-360.xlsx", skiprows=1)

# Muestra los datos leídos+
print(datos)
```

```python
"""
En este código, datos[datos['Mes'].str.contains('dic.')] filtra las filas donde la columna "Mes" contiene la cadena "dic.". 
Luego, pd.to_datetime() convierte la columna "Mes" a tipo datetime, como se explicó en la respuesta anterior.

"""

# Filtra las filas que contienen "dic." en la columna "Mes"
datos_dic = datos[datos['Mes'].str.contains('dic.')]

# Muestra los datos filtrados
print(datos_dic) # datos _dic es nuestro nuevo datafram
```

```python
"""
vamos ahora a filtrar hasta 2019

"""

# Selecciona las filas desde 1994 hasta 2019
datos_1994_2019 = datos_dic.loc[(datos_dic['Mes'] >= 'dic. 1994') & (datos_dic['Mes'] <= 'dic. 2019')]

```

```python
datos_1994_2019
```

```python
datos_1994_2019
```

```python
# ahora vamos a eliminar ese molesto "dic." de una vez por todas lo que vamos a hacer es 

# Elimina "dic." de la columna "Mes" y deja solo el año
datos_1994_2019['Mes'] = datos_1994_2019['Mes'].str.replace('dic. ', '')
```

```python
datos_1994_2019
```

```python
datos_1994_2019.info()
```

```python
# Convertir la columna "Mes" a tipo entero
datos_1994_2019['Mes'] = datos_1994_2019['Mes'].str[-4:].astype(int)

# Convertir la columna "Precio" a tipo entero
datos_1994_2019['Precio'] = datos_1994_2019['Precio'].str.replace(',', '.').astype(float).astype(int)

# Muestra los tipos de datos después de la conversión
print(datos_1994_2019.dtypes)
```
