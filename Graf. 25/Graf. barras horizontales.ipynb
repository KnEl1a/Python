```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Datos
df3 = pd.DataFrame({
    'Año': [2015, 2016, 2017, 2018, 2019],
    'Arg': [12.49, 10.96, 8.13, 6.06, 9.45],
    'Bra': [84.87, 86.84, 89.02, 91.75, 87.88],
    'Par': [0.07, 0.05, 0.05, 0.04, 0.01],
    'Uru': [2.58, 2.15, 2.79, 2.15, 2.65]
})

species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del DataFrame df3
years = df3['Año'].values
countries = df3.columns[1:].tolist()
data = df3.iloc[:, 1:].values.T

# Configuración de los colores
colors = plt.cm.tab10.colors[:len(countries)]

# Ancho de las barras
width = 0.6

# Crear el gráfico
fig, ax = plt.subplots()

# Lista para almacenar las barras
bars = []

# Bucle para agregar las barras apiladas
for i, country in enumerate(countries):
    if i == 0:
        bar = ax.bar(years, data[i], width, label=country, color=colors[i])
    else:
        bar = ax.bar(years, data[i], width, bottom=np.sum(data[:i], axis=0), label=country, color=colors[i])
    bars.append(bar)

# Añadir etiquetas con los valores en el centro de cada barra
for bar in bars:
    for rect in bar:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2, height / 2, f'{height:.2f}', ha='center', va='center')

# Configuraciones adicionales del gráfico
ax.set_title('Religiosidad en Sudamérica por Año')
ax.set_xlabel('Año')
ax.set_ylabel('Porcentaje de la población')
ax.legend()

# Mostrar el gráfico
plt.show()
```

```python
#----------


import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame con los datos
df3 = pd.DataFrame({
    'Año': [2015, 2016, 2017, 2018, 2019],
    'Argentina': [12.49, 10.96, 8.13, 6.06, 9.45],
    'Brasil': [84.87, 86.84, 89.02, 91.75, 87.88],
    'Paraguay': [0.07, 0.05, 0.05, 0.04, 0.01],
    'Uruguay': [2.58, 2.15, 2.79, 2.15, 2.65]
})

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))

# Configurar el estilo y los colores
sns.set(style="whitegrid")
colors = sns.color_palette("husl", n_colors=len(df3.columns) - 1)

# Definir el orden de las columnas para apilarlas
stacked_data = df3.set_index('Año').stack().reset_index().rename(columns={0: 'Porcentaje', 'level_1': 'País'})

# Crear el gráfico
sns.barplot(x='Porcentaje', y='Año', hue='País', data=stacked_data, palette=colors)

# Añadir etiquetas con los valores en el centro de cada barra
for index, row in stacked_data.iterrows():
    plt.text(row['Porcentaje'], index % 5, f'{row["Porcentaje"]:.2f}', ha='center', va='center')

plt.title('Religiosidad en Sudamérica por Año')
plt.xlabel('Porcentaje de la población')
plt.ylabel('Año')
plt.legend(title='País')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Especies de pingüinos
species = ('Adelie', 'Chinstrap', 'Gentoo')

# Conteo de pingüinos por sexo
sex_counts = {
    'Male': np.array([73, 34, 61]),   # Cantidad de pingüinos macho por especie
    'Female': np.array([73, 34, 58])  # Cantidad de pingüinos hembra por especie
}

# Ancho de las barras
width = 0.6

# Crear una figura y ejes
fig, ax = plt.subplots()

# Inicializar la posición inferior de las barras apiladas
bottom = np.zeros(3)

# Iterar sobre los conteos de pingüinos por sexo
for sex, sex_count in sex_counts.items():
    # Crear las barras para cada especie
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    
    # Actualizar la posición inferior para las próximas barras
    bottom += sex_count

    # Añadir etiquetas en el centro de cada barra
    ax.bar_label(p, label_type='center')

# Configurar el título del gráfico
ax.set_title('Number of penguins by sex')

# Añadir una leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```
