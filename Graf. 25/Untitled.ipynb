```python
import seaborn as sns
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

# Especies de pingüinos
species = ('Adelie', 'Chinstrap', 'Gentoo')

# Conteo de pingüinos por sexo
sex_counts = {
    'Male': np.array([73, 34, 61]),   # Cantidad de pingüinos macho por especie
    'Female': np.array([73, 34, 58])  # Cantidad de pingüinos hembra por especie
}

# Ancho de las barras
width = 0.6

# Crear una figura y ejes
fig, ax = plt.subplots()

# Inicializar la posición inferior de las barras apiladas
bottom = np.zeros(3)

# Iterar sobre los conteos de pingüinos por sexo
for sex, sex_count in sex_counts.items():
    # Crear las barras para cada especie
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    
    # Actualizar la posición inferior para las próximas barras
    bottom += sex_count

    # Añadir etiquetas en el centro de cada barra
    ax.bar_label(p, label_type='center')

# Configurar el título del gráfico
ax.set_title('Number of penguins by sex')

# Añadir una leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```

```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
    'Ojo al piojo': (167.95, 150.82, 190.19)
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
    'Ojo al piojo': (167.95, 150.82, 190.19)
}

x = np.arange(len(species))  # the label locations
total_attributes = len(penguin_means)
width = 0.8 / total_attributes  # Determina el ancho de cada barra y ajusta según el número de atributos
space = 0.1  # Espacio entre las barras de diferentes atributos
fig, ax = plt.subplots()

for i, (attribute, measurement) in enumerate(penguin_means.items()):
    offset = (width + space) * i - (total_attributes - 1) * (width + space) / 2
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x)
ax.set_xticklabels(species)
ax.legend(loc='upper left', bbox_to_anchor=(0, 1), ncol=2)  # Ajusta la ubicación de la leyenda
ax.set_ylim(0, 250)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
    'Ojo al piojo': (167.95, 150.82, 190.19)
}

x = np.arange(len(species))  # the label locations
width = 0.2  # the width of the bars
space_after_group = 0.5  # the space after each group of bars
total_attributes = len(penguin_means)
group_width = (width * total_attributes) + (space_after_group * ((total_attributes - 1) // 4 + 1))
fig, ax = plt.subplots()

for i, (attribute, measurement) in enumerate(penguin_means.items()):
    offset = (i % 4) * (width + space_after_group) + (i // 4) * space_after_group
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + group_width / 2 - width / 2)
ax.set_xticklabels(species)
ax.legend(loc='upper left', bbox_to_anchor=(0, 1), ncol=2)  # Adjusting the legend location
ax.set_ylim(0, 250)

plt.show()
```
