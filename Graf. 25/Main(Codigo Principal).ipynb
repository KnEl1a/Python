```python
import seaborn as sns
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt

df4 = pd.read_excel(r"C:\Users\Usuario\Downloads\Participación de cada uno de los países del Mercosur en las exportaciones del bloque a China 2015-2019.xlsx")
df = df4[['Unnamed: 0','Arg', 'Bra', 'Par', 'Uru']].round(2)
```

```python
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FuncFormatter  # Importa FuncFormatter

plt.figure(figsize=(10, 5))

# Establecer límites de los ejes x e y con un margen
#plt.xlim(right=150)  # Ajusta el límite derecho del eje x

# Modificado
a = plt.barh(df['Unnamed: 0'], df['Arg'], label="Argentina", color = "blue")
b = plt.barh(df['Unnamed: 0'], df['Bra'], left=df['Arg'], label="Brasil", color = "green")
c = plt.barh(df['Unnamed: 0'], df['Par'], left=df['Arg'] + df['Bra'], label="Paraguay", color = "red")
d = plt.barh(df['Unnamed: 0'], df['Uru'], left=df['Arg'] + df['Bra'] + df['Par'], label="Uruguay", color = "yellow")

#plt.title("Religiosidad en Sudamérica")
plt.xlabel("Porcentaje de participacion en las exportaciones a China")
plt.ylabel("Año")
plt.bar_label(a, label_type='center',fmt='%.2f%%')
plt.bar_label(b, label_type='center',fmt='%.2f%%')
plt.bar_label(c, label_type='center', padding=-14.9,fmt='%.2f%%')
plt.bar_label(d, label_type='center', padding=23.5,fmt='%.2f%%')

# Eliminar los números del eje x
plt.xticks([])

# Establecer los ticks del eje y
plt.yticks(ticks=df['Unnamed: 0'])

# Rotar las etiquetas del eje y para que estén en posición horizontal
plt.gca().invert_yaxis()

# Eliminar el marco alrededor del gráfico
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

#plt.legend()
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=4)
#plt.savefig(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 25\Graf. Baras H (1).png', dpi=300, bbox_inches='tight')
plt.show()
```
