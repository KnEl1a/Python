```python
# Data Manipulation
import pandas as pd
import openpyxl as op

# Data Visualisation
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# Importing dataset
df = pd.read_excel('Datos 6.xlsx')

df
# una hoja de calculo
```

```python
df_nuevo = df.rename(columns={'Import (US$ Thousand)': 'Importaciones (US$ Miles)'}, inplace=True)

df.rename(columns={'Partner Name': 'País'}, inplace=True)

df
```

```python
# Seleccionar la columna deseada para el Top 10
columna_a_considerar = 'Importaciones (US$ Miles)'

# Crear un nuevo DataFrame con el Top 10 basado en la columna seleccionada
top_10_df = df.nlargest(10, columna_a_considerar)

top_10_df
```

```python
top_10_df['Importaciones (US$ Miles)'] = top_10_df['Importaciones (US$ Miles)'].round().astype(int)

print(top_10_df.shape)
top_10_df
```

```python
# Cambiar nombres de filas para la columna 'Paises'
# Cambiar nombres de filas para la columna 'País'
nuevos_nombres_filas = {'Brazil': 'Brasil', 'Russian Federation': 'Rusia', 'United States': 'EEUU'}
top_10_df['País'].replace(nuevos_nombres_filas, inplace=True)

top_10_df

nuevonombre2= {'Thailand':'Tailandia', 'New Zealand':'Nueva Zelanda'}

top_10_df['País'].replace(nuevonombre2, inplace=True)

top = top_10_df.rename(columns ={'Importaciones (US$ Miles)':'Importaciones (US$)'})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Establecer el estilo de Seaborn
sns.set(style="whitegrid")

# Crear un gráfico de barras horizontales
plt.figure(figsize=(10, 6))  # Ajusta el tamaño del gráfico según tus preferencias

# Utilizar el método barplot de Seaborn
# Selecciona una paleta de colores con el parámetro 'palette'
ax = sns.barplot(x='Importaciones (US$)', y='País', data=top_10_df, palette='viridis')

# Ajustar el formato del eje x para mostrar números enteros
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))

# Ajustar el formato del eje y para mostrar números enteros sin notación científica
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))

# Guardar la figura en formato PNG con resolución de 300 ppp (píxeles por pulgada)
plt.savefig('grafico_punto_6.png')

# Mostrar el gráfico
plt.show()
```

```python
# Visualizar el DataFrame como tabla con matplotlib 2009
fig, ax = plt.subplots(figsize=(8,8))
ax.axis('tight')
ax.axis('off')
ax.table(cellText=top_10_df.values, colLabels=top_10_df.columns, cellLoc='center', loc='center')

# Guardar la visualización como una imagen PNG
plt.savefig('tabla pto 6.png', bbox_inches='tight')

plt.show()
```
