En este código, `plt.savefig` se utiliza para guardar el gráfico como un archivo PNG. El parámetro `dpi` controla la resolución de la imagen (puntos por pulgada). Aumentar el valor de `dpi` aumentará la calidad de la imagen, pero también el tamaño del archivo.

El parámetro `bbox_inches='tight'` se utiliza para asegurarse de que todos los elementos del gráfico se guarden en la imagen sin ser recortados.

Después de ejecutar este código, encontrarás el archivo PNG en el directorio de trabajo con el nombre especificado. Puedes cambiar `'nombre_del_archivo.png'` a la ruta y nombre que desees. Luego, puedes insertar este archivo PNG de alta calidad en tu documento de Word.

````python
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["225 g flour",
          "90 g sugar",
          "1 egg",
          "60 g butter",
          "100 ml milk",
          "1/2 package of yeast"]

data = [225, 90, 50, 60, 100, 5]

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f"angle,angleA=0,angleB={ang}"
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

#ax.set_title("Matplotlib bakery: A donut")

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Supongamos que ya tienes df_2000 definido
df_2000 = pd.read_excel(r"C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 11\Exportaciones MERCOSUR 2000 edit.xlsx")

labels = df_2000['Category']
sizes = df_2000['FOB US$']

# Ajustar el tamaño de la figura
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))

wedges, texts, autotexts = ax.pie(sizes, autopct='%1.1f%%', textprops=dict(color="w"), startangle=50,
                                  wedgeprops=dict(width=0.5, linewidth=0.4, edgecolor='black'), pctdistance=0.89)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"), bbox=bbox_props, zorder=0, va="center")

for i, (p, label) in enumerate(zip(wedges, labels)):
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f"angle,angleA=0,angleB={ang}"
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    
    # Ajustar la posición de las etiquetas y usar color de texto automático
    ax.annotate(label, xy=(x, y), xytext=(1.35 * np.sign(x), 1.4 * y),
                horizontalalignment=horizontalalignment, **kw, color=p.get_facecolor())

# Puedes personalizar la visualización según tus necesidades
# ax.set_title("Matplotlib bakery: A donut")
plt.savefig(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 11\dona año2000.png', dpi=300, bbox_inches='tight')
plt.show()
````
