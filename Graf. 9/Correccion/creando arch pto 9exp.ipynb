```python
import pandas as pd

# Lista de archivos
archivos = ['C:\\Users\\Usuario\\Downloads\\bra export.csv',
            'C:\\Users\\Usuario\\Downloads\\parag export.csv',
            'C:\\Users\\Usuario\\Downloads\\urug export.csv']

# Lista para almacenar DataFrames transformados
dfs_transformados = []

# Iterar sobre los archivos
for archivo in archivos:
    # Leer el archivo CSV y cargar los datos en un DataFrame
    df = pd.read_csv(archivo, delimiter=';', decimal=',')
    
    # Transformar la columna 'Exportaciones' a enteros
    df['Exportaciones'] = df['Exportaciones'].apply(lambda x: int(x) if x == x else x)
    
    # Agregar el DataFrame transformado a la lista
    dfs_transformados.append(df)

# Concatenar los DataFrames en uno solo
df_final = pd.concat(dfs_transformados, axis=1)

# Mostrar el DataFrame resultante
print(df_final)
```

```python
import pandas as pd

# Lista de archivos
archivos = ['C:\\Users\\Usuario\\Downloads\\bra export.csv',
            'C:\\Users\\Usuario\\Downloads\\parag export.csv',
            'C:\\Users\\Usuario\\Downloads\\urug export.csv']

# Lista para almacenar Series de Exportaciones transformadas
series_transformadas = []

# Iterar sobre los archivos
for archivo in archivos:
    # Leer el archivo CSV y cargar los datos en un DataFrame
    df = pd.read_csv(archivo, delimiter=';', decimal=',')
    
    # Transformar la columna 'Exportaciones' a enteros
    exportaciones_transformadas = df['Exportaciones'].apply(lambda x: int(x) if x == x else x)
    
    # Agregar la Serie transformada a la lista
    series_transformadas.append(exportaciones_transformadas)

# Concatenar las Series en una sola DataFrame
df_final = pd.concat(series_transformadas, axis=1)

# Mostrar el DataFrame resultante
print(df_final)

```

```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FuncFormatter  # Importa FuncFormatter

# Setting the figure size
plt.figure(figsize=(16, 5))

def format_years(x, pos):
    return int(x)

def format_y_plain(x, pos):
    return f'{x/1000000000:.0f}'


plt.bar(df_graf_exp['AÑOS'], df_graf_exp['arg_export'], width=1/5, label = "Argentina", color = "blue")
plt.bar(df_graf_exp['AÑOS'] + 0.2, df_graf_exp['bra_export'], width=1/5, label = "Brasil", color = "green")
plt.bar(df_graf_exp['AÑOS'] + 0.4, df_graf_exp['parag_export'], width=1/5, label = "Paraguay", color = "red")
plt.bar(df_graf_exp['AÑOS'] + 0.6, df_graf_exp['urug_export'], width=1/5, label = "Uruguay", color = "yellow")


plt.legend()

plt.xlabel("Años") #etiqueta en eje x
plt.ylabel("Total FOB US$ [Mil Millones]")#etiqueta en eje y


formatter = FuncFormatter(format_years)
plt.gca().xaxis.set_major_formatter(formatter)


y_formatter = FuncFormatter(format_y_plain)
plt.gca().yaxis.set_major_formatter(y_formatter)
plt.xticks(np.arange(2000, 2020, 1))


plt.minorticks_on()

plt.grid(True, which='major', axis='both', color='black', linestyle='--', linewidth=0.5, alpha=0.3)
plt.savefig('export_graf9.png', dpi=300, bbox_inches='tight')
plt.show()
```
