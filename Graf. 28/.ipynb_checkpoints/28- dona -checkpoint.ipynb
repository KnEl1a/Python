```python
import seaborn as sns
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt

df4 = pd.read_excel(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 28\Datos28.xlsx')
df = df4[['%']].round(2)
```

```python
df4
```

```python
df['Sector'] = df4['Sector']
```

```python
df
```

```python
# Eliminar las filas donde el valor de la columna "Sector" es "TOTAL"
df = df[df['Sector'] != 'TOTAL']
```

```python
df
```

```python
total = df['%'].sum()
```

```python
total
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Supongamos que ya tienes data definido
labels = df['Sector']
sizes = df['%']

# Ajustar el tamaño de la figura
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))

# Definir una paleta de colores roja
colors = plt.cm.Reds(np.linspace(0.2, 0.8, len(labels)))

wedges, texts, autotexts = ax.pie(sizes, autopct='%1.1f%%', textprops=dict(color="black"), startangle=50,
                                  wedgeprops=dict(width=0.5, linewidth=0.4, edgecolor='black'), pctdistance=0.925, colors=colors)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-", color="black"), bbox=bbox_props, zorder=0, va="center")

for i, (p, label) in enumerate(zip(wedges, labels)):
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f"angle,angleA=0,angleB={ang}"
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    
    # Ajustar la posición de las etiquetas y usar color de texto automático
    ax.annotate(label, xy=(x, y), xytext=(1.35 * np.sign(x), 1.4 * y),
                horizontalalignment=horizontalalignment, **kw, color="black")

# Puedes personalizar la visualización según tus necesidades
# ax.set_title("Matplotlib bakery: A donut")
plt.show()
```
