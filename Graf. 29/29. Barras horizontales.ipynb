```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Cargar los datos desde el archivo CSV
#data = pd.read_csv("principales-pases.csv", sep=';', decimal=',')
#data = pd.read_excel(r"C:\Users\Usuario\Downloads\Participación de cada uno de los países del Mercosur en las exportaciones del bloque a China 2015-2019.xlsx")
df = pd.read_excel(r"C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 29\Datos6.xlsx")
```

```python
df.head()
```

```python
data = df.sort_values(by='Monto (miles de millones $US)', ascending=True) # de menor a mayor
```

```python
data
```

```python
data = data[data['País'] != 'TOTAL']
```

```python
# Configurar la paleta de colores azul-celeste
#custom_palette = sns.color_palette(["red"])

# Crear el gráfico de barras horizontales
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Monto (miles de millones $US)", y="País", data=data, palette="Blues")

# Añadir título y etiquetas
#plt.title('Inversiones y contratos chinos en america latina ')
plt.xlabel('')
plt.ylabel('')

# Cambiar el formato del eje x a mil millones de dólares
#ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}'.format(x/1e9)))

# Agregar etiquetas con porcentaje dentro de cada barra
for p in ax.patches:
    ax.text(p.get_width() * 1.02, p.get_y() + p.get_height() / 2, '{:.2f}'.format(p.get_width()), ha='left', va='center')

# Eliminar el marco alrededor del gráfico
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)



plt.grid(True, which='major', axis='x', color='black', linestyle='--', linewidth=0.5, alpha=0.3)
#plt.savefig('grafico_exportaciones.png', dpi=300, bbox_inches='tight')
plt.savefig(r'C:\Users\Usuario\Desktop\Port Folio\Pandas y Matplotlib\Proyecto I+D UNO\Graf. 29\B_horizontales.png', dpi=300, bbox_inches='tight')
plt.show()
```
